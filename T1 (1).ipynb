```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sps
from random import *
```

```python
def estimation_1(n,mas):
    return 2*sum(mas[0:n])/n
```

```python
def estimation_2(n,mas):
    return (n+1)*min(mas[0:n])
```

```python
def estimation_3(n,mas):
    return (n+1)/n*max(mas[0:n])
```

```python
def estimation_4(n,mas):
    return min(mas)+max(mas[0:n])
```

```python
def estimation_5(n,mas):
    return mas[0]+1/(n-1)*sum(mas[1:n])
```

```python
N=100
T=np.random.randint(1,N)
value = sps.uniform.rvs(size=N)*T 
y_1 = abs(np.array([estimation_1(i, value) for i in range(1, len(value)+1)])-T)
y_2= abs(np.array([estimation_2(i, value) for i in range(1, len(value)+1)]))
y_3 = abs(np.array([estimation_3(i, value) for i in range(1, len(value)+1)])-T)
y_4 = abs(np.array([estimation_4(i, value) for i in range(1, len(value)+1)]) -T)
y_5 = abs(np.array([estimation_5(i, value) for i in range(2, len(value)+2)])-T)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter([0],[0], color = "white", label = f'teta = {T}')
ax.plot(np.arange(1, N+1), y_1, color = 'black', label= '2*(sum x_n)/n')
ax.plot(np.arange(1, N+1), y_3, color = 'yellow', label = '(n+1)/n*(max x_n)')
ax.plot(np.arange(1, N+1), y_4, color = 'red', label = '(max of x_n) + (min x_n)')
ax.plot(np.arange(1, N+1), y_5, color = 'blue', label = 'x_1 + 1/(n-1)*(sum x_2; x_n)')
ax.set_ylabel('teta* - teta')
ax.legend(loc='upper left', shadow=True, framealpha=1, edgecolor='black')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter([0],[0], color = "white", label = f'teta = {T}')
ax.plot(np.arange(1, N+1), y_2, color = 'green', label = '(n+1)*(min of X_n)')
ax.set_ylabel('teta* - teta')
ax.legend(loc='upper right', shadow=True, framealpha=1, edgecolor='black')
plt.show()
```
